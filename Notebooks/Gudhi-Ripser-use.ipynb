```python
import iblofunmatch.inter as ibfm
import numpy as np

RandGen = np.random.default_rng(5)
C0 = ibfm.sampled_circle(7,7.5,50, RandGen)
C1 = ibfm.sampled_circle(0,2.5,50, RandGen)
X = np.vstack((C0, C1))
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(X[:,0], X[:,1], c="red", zorder=2, marker="X", s=150)
ax.set_aspect("equal")
fig.tight_layout()
```

```python
from ripser import Rips
import time
start = time.time()
rips = Rips(maxdim=1)
start = time.time()
diagrams = rips.fit_transform(X)
print(f"Ripser took {time.time()-start} seconds")
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.scatter(diagrams[0][:,0], diagrams[0][:,1])
ax.scatter(diagrams[1][:,0], diagrams[1][:,1])
ax.plot([0,8], [0,8], color="gray")
```
